1️⃣ User Login Check

```python
username = "admin"
password = "1234"

if username == "admin" and password == "1234":
    print("Login Successful")
else:
    print("Invalid Credentials")
```

2️⃣ Pass / Fail Analyzer

```python
marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

for mark in marks:
    if mark >= 50:
        pass_count = pass_count + 1
    else:
        fail_count = fail_count + 1

print("Total Pass Students:", pass_count)
print("Total Fail Students:", fail_count)
```

3️⃣ Simple Data Cleaner

```python
names = [" Alice ", "bob", " CHARLIE "]

clean_names = []

for name in names:
    name = name.strip()
    name = name.lower()
    clean_names.append(name)

print("Cleaned Names:", clean_names)
```

4️⃣ Message Length Analyzer

```python
messages = ["Hi", "Welcome to the platform", "OK"]

for message in messages:
    length = len(message)
    print("Length of message:", length)

    if length > 10:
        print("Message is longer than 10 characters")
```

5️⃣ Error Message Detector

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = 0

for log in logs:
    if log == "ERROR":
        error_count = error_count + 1

print("Total ERROR count:", error_count)
```
