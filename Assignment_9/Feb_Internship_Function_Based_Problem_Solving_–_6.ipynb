# Smart Parking Lot Management System

This notebook implements a function to manage a parking lot.
It tracks vehicle entries and exits, calculates current vehicles,
identifies peak usage, and checks if capacity is exceeded.

```python
def manage_parking(capacity, vehicle_logs):
    """
    Calculates current parked vehicles and parking status.
    """

    current = 0

    # Count vehicles using loop
    for log in vehicle_logs:
        if log == "IN":
            current += 1
        elif log == "OUT":
            if current > 0:
                current -= 1

    # Determine status
    if current > capacity:
        status = "Full"
    else:
        status = "Available"

    print(f"Currently Parked Vehicles: {current}")
    print(f"Parking Status: {status}")
```

```python
capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

manage_parking(capacity, vehicle_logs)
```

# Online Food Delivery Time Estimator

This notebook creates a function that estimates food delivery time
based on distance, traffic level, and weather condition.
The function dynamically adjusts time using logical conditions.

```python
def estimate_delivery_time(distance, traffic, weather):
    """
    Estimates delivery time using distance and conditions.
    """

    # Base time (5 minutes per km)
    base_time = distance * 5

    # Traffic adjustment
    if traffic == "Low":
        traffic_delay = 5
    elif traffic == "Medium":
        traffic_delay = 10
    elif traffic == "High":
        traffic_delay = 10
    else:
        traffic_delay = 0

    # Weather adjustment
    if weather == "Rainy":
        weather_delay = 5
    else:
        weather_delay = 0

    total_time = base_time + traffic_delay + weather_delay

    print(f"Estimated Delivery Time: {total_time} minutes")
```

```python
estimate_delivery_time(8, "High", "Rainy")
```

# Movie Theatre Seat Occupancy Analyzer

This notebook defines a function that evaluates theatre seat bookings.
It calculates the occupancy percentage, determines show status,
and suggests whether additional shows may be required.

```python
def analyze_occupancy(total_seats, booked_seats):
    """
    Calculates occupancy and determines show status.
    """

    booked_count = len(booked_seats)

    occupancy = (booked_count / total_seats) * 100

    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 75:
        status = "Almost Full"
    else:
        status = "Seats Available"

    print(f"Occupancy: {int(occupancy)}%")
    print(f"Show Status: {status}")
```

```python
total_seats = 200
booked = [1] * 150

analyze_occupancy(total_seats, booked)
```

# Cloud Server Load Classification System

This notebook defines a function that analyzes CPU usage readings.
It calculates the average CPU load and classifies the server status
as Normal, Warning, or Critical based on defined thresholds.

```python
def classify_server_load(cpu_readings):
    """
    Calculates average CPU load and classifies server status.
    """

    total = 0

    # Calculate total using loop
    for value in cpu_readings:
        total += value

    average = total / len(cpu_readings)

    if average < 50:
        status = "Normal"
    elif average <= 80:
        status = "Warning"
    else:
        status = "Critical"

    print(f"Average CPU Load: {round(average)}%")
    print(f"Server Status: {status}")
```

```python
cpu_data = [45, 60, 70, 85, 90]

classify_server_load(cpu_data)
```

# Smart Classroom Resource Usage Monitor

This program analyzes classroom resource usage data.
It identifies overused resources and triggers an energy alert
based on predefined usage thresholds.

```python
def monitor_resources(resource_usage):
    """
    Identifies overused classroom resources.
    """

    threshold = 8
    overused = []

    # Check each resource using loop
    for resource, hours in resource_usage.items():
        if hours > threshold:
            overused.append(resource)

    if overused:
        print(f"Overused Resources: {', '.join(overused)}")
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")
```

```python
usage_data = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

monitor_resources(usage_data)
```

# Online Event Registration Capacity Controller

This program manages event registrations.
It ensures capacity limits are respected,
prevents overbooking, and moves extra users to a waitlist.

```python
def manage_registrations(capacity, registrations):
    """
    Manages event registrations and handles waitlist.
    """

    confirmed = 0
    waitlisted = 0

    # Process registrations using loop
    for _ in range(registrations):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1

    status = "Closed" if waitlisted > 0 else "Open"

    print(f"Confirmed Registrations: {confirmed}")
    print(f"Waitlisted Users: {waitlisted}")
    print(f"Registration Status: {status}")
```

```python
manage_registrations(100, 105)
```
